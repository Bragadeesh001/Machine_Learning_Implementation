# Ridge and Lasso Regression


```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing sklearn dataset
from sklearn.datasets import load_boston
```

```python
# loading boston

df=load_boston()
df
```

```python
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
dataset.head()
```

```python
dataset['Price']=df.target
dataset.head()
```

```python
# Splitting the dataset
X=dataset.drop(columns='Price')
y=dataset['Price']
```

```python
# implementing linear regression

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

regression=LinearRegression()
mse=cross_val_score(regression,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

This mean squared error score has to be nearer to zero

the more nearer to zero the more accurate the model is

```python
# implementing Ridge regression

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,40,45,50,55,100]}
ridge_regresser=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error', cv=5)
ridge_regresser.fit(X,y)


print(ridge_regresser.best_params_)
print(ridge_regresser.best_score_)
```

```python
# implementing lasso

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,40,45,50,55,100]}
lasso_regresser=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error', cv=5)
lasso_regresser.fit(X,y)

print(lasso_regresser.best_params_)
print(lasso_regresser.best_score_)
```

```python
# train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
prediction_ridge=ridge_regresser.predict(X_test)
prediction_lasso=lasso_regresser.predict(X_test)
```

```python
sns.displot(y_test-prediction_ridge,kde=True)
```

```python
sns.displot(y_test-prediction_lasso,kde=True)
```
